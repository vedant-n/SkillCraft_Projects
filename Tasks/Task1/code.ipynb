```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the data
data = pd.read_csv('population.csv')
```

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
data.columns
```

```python
# Select the variable you want to visualize
variable = "Year"  
```

```python
# Select the variables you want to plot
x = data["Year"]
y = data["World Population"]

# Create the area chart
plt.fill_between(x, y, alpha=0.5, label="World Population")

# Add labels and title
plt.xlabel("Year")
plt.ylabel("World Population")
plt.title("Year vs World Population")

# Show the legend
plt.legend()

# Show the plot
plt.show()

# Create a histogram for continuous variables
if data[variable].dtype.kind in 'bifc':  # check if the variable is numeric
    plt.hist(data[variable], bins=50, edgecolor='black')
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("Distribution of " + variable)
    plt.show()

# Create a bar chart for categorical variables
else:
    counts = data[variable].value_counts()
    plt.bar(counts.index, counts.values, edgecolor='black')
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("Distribution of " + variable)
    plt.show()
```
